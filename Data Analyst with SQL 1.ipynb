```python
# select all the data
ps.sqldf("""
    SELECT 
    df.* FROM ti AS df
""")
```

```python
# select Limit to get top 5

ps.sqldf("""
    SELECT 
    df.* FROM ti AS df
    LIMIT 5
""")
```

```python
# Taking only PassengerID and the name frpm dataset
ps.sqldf("""
    SELECT
    df.PassengerId,
    df.Name           
    FROM ti AS df
    LIMIT 5
""")
```

```python
# RENAME THE TWO CHOSEN COLUMN
ps.sqldf("""
SELECT 
    df.PassengerId AS ID_OF_Passenger,
    df.Name AS Name_Of_Passenger
    FROM ti AS df
    LIMIT 5
""")
```

```python
# Sele t all the passenger whoes age is below 1 year
ps.sqldf("""
    SELECT 
    df.*
    FROM ti AS df
    WHERE df.age < 1
""")
```

```python
# Giving an alias name to the dataset and using it in "SELECT" statement too

ps.sqldf("""

SELECT
    df.*
    FROM ti as df
    
""")
```

```python
# You can check for missing value by using "ISNULL"

ps.sqldf("""
SELECT 
    df.*
    FROM ti as df
    WHERE df.age NOT NULL
""")
```

```python
# Example for "BETWEEN" keyword

ps.sqldf("""
SELECT 
df.*
FROM ti AS df
WHERE df.age BETWEEN 5 AND 10
""")
```

```python
# Selecting everything from the Dataset for passenger whoes age is beloww 1 year and who are females

ps.sqldf("""
    SELECT
    df.*
    FROM ti AS df
    WHERE df.age<1
    AND df.Sex="female"
""")
```

```python
# Selecting everything from the Dataset for passenger whoes age is beloww 1 year and who are females or age is above 70 year and Pclass is 1

ps.sqldf("""
    SELECT df.*
    FROM ti AS df
    WHERE (df.Age<1 AND df.Sex="female")
    OR (df.Age>70 AND df.Pclass=1)
""")
```

```python
# selecting everthing from the dataset for passenger where cabin is one og A5,A6, A7

ps.sqldf("""
SELECT df.*
FROM ti AS df
WHERE df.Cabin IN ('A5','A6','A7')
""")
```

```python
# Selecting everything from the datasets for passengers where Age is one of 65,70,71

ps.sqldf("""
SELECT df.*
FROM ti AS df
WHERE df.Age IN (65,70,71)
""")
```

```python
# selecting everything from the datasets for passenger where name contains 'Mrs.'

ps.sqldf("""
SELECT df.*
FROM ti AS df
WHERE df.Name LIKE "%MRS.%"
LIMIT 5
""")
```

```python
# Adding a column for Square,Log,Root etc of
ps.sqldf("""
SELECT 
df.Age,
df.Age * df.Age AS age_sq,
df.Age + 1 AS df.plus
FROM ti AS df
LIMIT 5

""")
```

```python
# concatenate two string
ps.sqldf("""
SELECT 
df.Sex,
df.Ticket,
df.Sex || df.Ticket AS Sex_plus_ticket
FROM ti AS df
LIMIT 5
""")
```

```python
# Using CASE WHEN to create buckets for age

ps.sqldf("""
SELECT 
df.Name,
df.PassengerID,
df.Age,
CASE 
WHEN df.Age IS NULL THEN '0. Missing'
WHEN df.Age < 18 THEN '1. 1-17'
WHEN df.AGE < 60 THEN '2. 18-60'
ELSE '3.60+' END AS age_bucket
FROM ti AS df
LIMIT 50

""")
```

```python
# Using SUBSTR() to take out a substring

ps.sqldf("""
SELECT 
df.Name,
SUBSTR(df.Name,2,4) AS out
FROM ti AS df
LIMIT 5
""")
```

```python
# Lower() and Upper()
ps.sqldf("""
Select
df.Name,
LOWER(df.Name) AS lcase_name,
UPPER(df.Name) AS ucase_name
FROM ti AS df
LIMIT 5
""")
```

```python
# Selecting all distinct vbalue of "Embarked" column
ps.sqldf("""
SELECT
DISTINCT df.Embarked
FROM ti AS df
""")
```
